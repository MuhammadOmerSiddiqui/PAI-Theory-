```python
import pandas as pd
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
reportLines = []
for column in df.columns:
    dataType = df[column].dtype
    missingPart = df[column].isna().mean() * 100
    reportLines.append(f'{column}: {dataType}, missing:{missingPart}')
with open("inspect_report.txt","w") as file:
    for line in reportLines:
        file.write(line+"\n")
```

```python
df["Age"] = df.groupby(['Pclass','Sex'])["Age"].transform(lambda g : g.fillna(g.median()))
embarkedMode = df["Embarked"].mode()[0]
df["Embarked"] = df["Embarked"].fillna(embarkedMode)
if("Cabin" in df.columns):
    df = df.drop(columns=["Cabin"])
```

```python
df['FamilySize'] = df["SibSp"] + df["Parch"]
df["IsAlone"] = (df["FamilySize"] == 0).astype(int)
df["Age"] = df["Age"].astype('int64')
```

```python
df.to_csv("titanic_cleaned.csv",index=False)
```
