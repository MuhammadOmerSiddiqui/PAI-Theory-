```python
import numpy as np
```

```python
my_list = []
with open("sensor_data.csv","r") as file:
    for item in file:
        row = item.strip().split(",")
        rowFloat = [float(value) for value in row]
        my_list.append(rowFloat)
```

```python
Array = np.array(my_list)
```

```python
Array[Array == -999] = np.nan
Array[(Array<0) | (Array>100)] = np.nan
```

```python
columnsMean = np.nanmean(Array,axis=0)
rowMedian = np.nanmedian(Array,axis=1)
nanCount = np.sum(np.isnan(Array),axis=0)
sensorMostInvalid = np.argmax(nanCount)
maxInvalidCount = nanCount[sensorMostInvalid]
```

```python
maxVal = np.nanmax(Array)
minVal = np.nanmin(Array)
dataNormalised = (Array - minVal) / (maxVal - minVal)
```

```python
np.savetxt("sensor_data_normalized.csv",dataNormalised,delimiter=",",fmt='%.6f')
```
